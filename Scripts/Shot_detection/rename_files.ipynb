```python
import os
import random
import string

def generate_random_name(length=8):
    """Generate a random string of letters and digits."""
    chars = string.ascii_letters + string.digits
    return ''.join(random.choice(chars) for _ in range(length))

def randomly_rename_files(folder_path, name_length=8):
    # List all files in the directory
    
    handsUp = os.listdir('video 1/Hands_Up')
    xml = os.listdir('video 1/XML_Files')
    # Separate image and text files
    image_extensions = {'.jpg', '.jpeg', '.png'}
    image_files = [f for f in handsUp if os.path.splitext(f)[1].lower() in image_extensions]
    text_files = [f for f in xml if f.endswith('.xml')]
    # Ensure images and text files have corresponding pairs
    image_file_names = set(os.path.splitext(f)[0] for f in image_files)
    text_file_names = set(os.path.splitext(f)[0] for f in text_files)
    common_names = image_file_names.intersection(text_file_names)
    # print(image_file_names)
    if not common_names:
        print("No matching image-text pairs found.")
        return
    # Rename files with random names
    for base_name in common_names:
        new_name = generate_random_name(name_length)
        
        old_image_path_jpg = os.path.join('video 1/Hands_Up', f"{base_name}.jpg")
        
        old_text_path = os.path.join('video 1/XML_Files', f"{base_name}.xml")
        
        # Rename image file if it exists
        if os.path.exists(old_image_path_jpg):
            new_image_path = os.path.join('video 1/Hands_Up', f"{new_name}.jpg")
            os.rename(old_image_path_jpg, new_image_path)
        
        # Rename text file
        if os.path.exists(old_text_path):
            new_text_path = os.path.join('video 1/XML_Files', f"{new_name}.xml")
            os.rename(old_text_path, new_text_path)
        
        print(f"Renamed {base_name} to {new_name}")

# Path to the folder containing images and text files
folder_path = 'video 1/'

# Run the rename function
randomly_rename_files(folder_path)
```

```python
image_files = os.listdir("video 1/Hands_Up")

label_dir = os.listdir("video 1/XML_Files")
missing_labels = []

for img_file in image_files:
    label_file = img_file.split('.')[0] + '.xml'

    if label_file not in label_dir:
        missing_labels.append(label_file)

if missing_labels:
    print(f"Missing label files for the following images: {missing_labels}")
else:
    print("All images have corresponding label files.")
```

```python
import os
import xml.etree.ElementTree as ET
def update_xml_file(xml_path, new_filename, new_path):
    tree = ET.parse(xml_path)
    root = tree.getroot()
    root.find('filename').text = new_filename
    root.find('path').text = new_path
    tree.write(xml_path)
    print(f"Updated {xml_path}: filename to {new_filename} and path to {new_path}")

xmlFiles=os.listdir("video 1/XML_Files")
for xml in xmlFiles:
    xml_path=f"video 1/XML_Files/{xml}"
    new_filename=xml
    new_path=f"Output\XML_Files\\{new_filename}"
    update_xml_file(xml_path,new_filename,new_path)
```
